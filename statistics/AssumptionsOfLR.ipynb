# 5 Assumptions of Linear Regression

Linear Regression cannot be blindly applied to all datasets.

1. Linearity: There is a linear relationshio between target variable and all independed variables.
2. Homoscedasticity (Equal Variance): We dont want a cone shaped plot, which means variance is dependent on independent variable.
3. Multivariate Normality (Normality of Error Distribution): Errors are normally distributed
4. No Auto Corelation (Independence): No patterns in the data
5. Lack of Multicollinearity: Predictors are not corelated with each other

6. Outlier Check: Consider removing outliers from the data
